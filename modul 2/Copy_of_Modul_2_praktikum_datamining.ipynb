```
#import librari pandas
import numpy as np
import pandas as pd
#import dataset dari csv/excel menjadi pandas
df=pd.read_csv('/content/iris.csv', sep=',',header=None)
#ubah dataset dari pandas ke array numpy

dataset = df.to_numpy()
#menghilangkan kolom dan baris Nan/kosong
dataset = dataset[:,3:8]
print(dataset[:5])
# type(dataset)
# dfdataset = pd.DataFrame(dataset, columns = ['Petal_width','Petal_length','Sepal_width','Sepal_length','Species_name'])
# dfdataset
```

```
#kelas setosa
setosa = np.array([isibaris[:4] for isibaris in dataset if isibaris[4] == ' Setosa'])
setosa = setosa.astype(np.float)
print(setosa[:5])
```

```
#kelas versicolor
versicolor = np.array([isibaris[:4] for isibaris in dataset if isibaris[4] == ' Versicolor'])
versicolor = versicolor.astype(np.float)
print(versicolor[:5])
```

```
#kelas virginica
virginica = np.array([isibaris[:4] for isibaris in dataset if isibaris[4] == ' Verginica'])
virginica = virginica.astype(np.float)
print(virginica[:5])
```

```
#kelas unknown
unknown = np.array([isibaris[:4] for isibaris in dataset if isibaris[4] == 'unknown'])
unknown = unknown.astype(np.float)
print(unknown[:3])
```

###8 Buatlah pararrel coordinat plot dari data yang berkelas setosa, versicolor, dan virginica. Berikan warna merah untuk setosa, hijau untuk versicolor, biru untuk virginica. Hint gunakan perintah plot dan operasi slicing pada numpy untuk mempermudah tugas anda 


```
import matplotlib.pyplot as plt
import numpy as np

#font
font1 = {'family':'serif','color':'black','size':20}
#judul, keterangan
plt.title("pararrel coordinat plot", fontdict = font1)

# setosa
for isibaris in setosa:
  y = np.array(isibaris)
  x = np.array(['Petal_width','Petal_length','Sepal_width','Sepal_length'])
  plt.plot(x, y, color = 'red')
#Versicolor
for isibaris in versicolor:
  y = np.array(isibaris)
  x = np.array(['Petal_width','Petal_length','Sepal_width','Sepal_length'])
  plt.plot(x, y, color = 'green')
#Verginica
for isibaris in virginica:
  y = np.array(isibaris)
  x = np.array(['Petal_width','Petal_length','Sepal_width','Sepal_length'])
  plt.plot(x, y, color = 'blue')

plt.show()
```
